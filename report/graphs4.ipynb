```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob
import csv
```

```python
def read_results(pattern):
    data = pd.DataFrame({})
    for file in glob.glob(pattern):
        print(file)
        raw = pd.read_csv(file, skipfooter=2)
        if (data.empty):
            data = raw
            print(len(raw))
        else:
            data = data.merge(raw, left_index=True, right_index=True, how='outer')
    return data

optimum_price = 3730

def get_stats(pattern):
    for file in glob.glob(pattern):

        with open(file,"r") as f:
            reader = csv.reader(f,delimiter = ",")
            data = list(reader)
            row_count = len(data)

        df = pd.read_csv(file, skiprows = row_count - 2)
        print(file)        
        print(df)
        print("-------")
```

```python
ax = read_results("../csvs/genetic/elitism_count*")
# bx = get_stats("../csvs/genetic/elitism_count*")
```

```python
ax
ax = ax.loc[:, ax.columns.str.contains('best_combination')]
ax = ax.reindex(sorted(ax.columns), axis=1)
plot= ax.plot(title="Elitism count")
plot.axhline(y=optimum_price, xmin=-1, xmax=1, color='limegreen', linestyle='--', lw=1)
```

```python
ax = read_results("../../csvs/genetic/elitism_count*")
ax = ax.loc[:, ax.columns.str.contains('best_combination')]
ax = ax.reindex(sorted(ax.columns), axis=1)
plot= ax.plot(title="Elitism count")
plot.axhline(y=optimum_price, xmin=-1, xmax=1, color='limegreen', linestyle='--', lw=1)
plot.set_xlabel("Generations")
plot.set_ylabel("Best price of item combinations")
plt.savefig('graphs/elitism_count', bbox_inches='tight')
get_stats("../../csvs/genetic/elitism_count*")
```
