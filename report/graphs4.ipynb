```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob
import csv
```

```python
def read_results(pattern):
    data = pd.DataFrame({})
    for file in glob.glob(pattern):
        print(file)
        raw = pd.read_csv(file, skipfooter=2)
        if (data.empty):
            data = raw
            print(len(raw))
        else:
            data = data.merge(raw, left_index=True, right_index=True, how='outer')
    return data

optimum_price = 3730

def get_stats(pattern):
    for file in glob.glob(pattern):

        with open(file,"r") as f:
            reader = csv.reader(f,delimiter = ",")
            data = list(reader)
            row_count = len(data)

        df = pd.read_csv(file, skiprows = row_count - 2)
        print(file)
#         print(type(df))
        print(df.to_latex())
#         df.describe()
        print("-------")
```

```python
type = "elitism_count"
ax = read_results("../csvs/genetic/{}*".format(type))
ax = ax.loc[:, ax.columns.str.contains('best_combination')]
ax = ax.reindex(sorted(ax.columns), axis=1)
plot= ax.plot(title="{}".format(type))
plt.savefig('graphs/{}'.format(type), bbox_inches='tight')

bx = get_stats("../csvs/genetic/{}*".format(type))
```
